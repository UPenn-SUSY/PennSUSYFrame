# Contribution plots
In this document I make several plots of the number of expected events and expected sensitivity in the proposed signal regions as a function of different cut values.

```python
import sys
import ContributionsHelpers as helpers
import pandas
import itertools
import matplotlib.pyplot as plt
%matplotlib inline

from ROOT import RooStats
```

## Read input from file
For this, I use a function defined elsewhere, the details aren't so important. I am simply grabbing the expected number of events, and raw number of events for each background in each region from our file of histograms

```python
input_file_name = '~/Desktop/scratch/compare_plots.b_minus_l.w_zkfactor.all_sig.root' ## TODO hard coded for now make configurable
region_df = helpers.extractRegionContributions(input_file_name)
region_df = helpers.cleanDataFrame(region_df)
```

The data frame has the following structure

```python
region_df.head()
```

## Pruning the data frame
This data frame has a lot of information we don't really need. Here, I extract the information relevant to this study.
- Drop all information from regions other than the signal regions
- Keep only information about:
    - $h_T$ and $m_{b\ell}$ cut values
    - Number of expected signal events
    - Number of raw MC signal events
    - Expected $Z_N$ significance
    - Backround expectation and uncertainty in the region

```python
# Drop information for regions other than the proposed SRs
subset = region_df[(region_df['region'].str.contains('SR HT'))]
subset.head()
```

```python
# Extract the expected background contribution and uncertainty in each region
region_contributions = helpers.getTotalRegionContribution(subset,
                                                          subset[subset['sample'].str.contains('stop')
                                                                 == False]['sample'].unique(),
                                                          subset['region'].unique())
region_contributions.head()
```

```python
# Create a new data frame to hold the results in the SR
result_df = pandas.DataFrame(columns = ('ht cut', 'mbl cut',
                                        'sample', 'count',
                                        'raw', 'zn',
                                        'bkg total', 'bkg uncertainty'))

# step through each sample and region combination in the subset data frame
for sample, region in itertools.product(subset[subset.sample.str.contains('stop')].sample.unique(),
                                        subset['region'].unique()):
    # extract the ht and mbl cut values frome the region name
    ht_cut = region.split()[2]
    mbl_cut = region.split()[4]

    # get the number of expected events and the raw number of events
    count = subset[(subset['sample'] == sample) &
                   (subset['region'] == region)].iloc[0]['count']
    raw = subset[(subset['sample'] == sample) &
                 (subset['region'] == region)].iloc[0]['raw']
    
    # extract the expected background events and the uncertainty
    bkg_total = region_contributions[region_contributions.region
                                     == region].iloc[0]['total']
    bkg_uncert = region_contributions[region_contributions.region
                                      == region].iloc[0]['uncertainty']

    # finally, get the Zn value
    zn = RooStats.NumberCountingUtils.BinomialExpZ(count,
                                                   bkg_total,
                                                   (bkg_uncert/bkg_total))
    
    # stick all this information into the resulting data frame
    result_df.loc[result_df.shape[0]] = [int(ht_cut),
                                         int(mbl_cut),
                                         sample,
                                         count,
                                         raw,
                                         zn,
                                         bkg_total,
                                         bkg_uncert]

result_df.head()
```

## Interpretation
Now that all the information has been extracted, it is possible to make plots and evaluate the signal region choices.

### $Z_{N}$ vs $h_T$ cut value

```python
def plotZn(df, mass):
    """
    Function to take a data frame (df), and a mass, and draw a plot
    of Zn vs the hT cut value for different choices of mbl cut value.
    Each mbl cut value is a line on the plot
    """
    # Create a copy of the data frame, with only the signal samples
    # we can drop several columns while we are at it
    zn = df[df.sample.str.contains(str(mass))][['ht cut',
                                                'mbl cut',
                                                'zn']]
    # get a list of the mbl cut values included in the data frame
    mbl_cut_vals = zn['mbl cut'].unique()
    
    # list of new column names to add to the data frame used
    # for plotting
    new_cols = ['ht cut']
    new_cols.extend([' '.join(['$m_{bl}$ >',
                               str(mbl),
                               'GeV']) for mbl in mbl_cut_vals])
    df_to_print = pandas.DataFrame(columns=new_cols)
    
    # step through the ht cut values, and add a new row
    # for each cut value
    for ht in zn['ht cut'].unique():
        to_add = [ht]
        to_add.extend([zn[(zn['ht cut'] == ht) &
                          (zn['mbl cut'] == mbl)].iloc[0]['zn']
                       for mbl in mbl_cut_vals])
        df_to_print.loc[df_to_print.shape[0]] = to_add
    
    # set the ht cut column to be the index of the data
    # frame
    df_to_print = df_to_print.set_index('ht cut').sort()
    
    # draw plot of Zn vs cut value
    ax = df_to_print.plot()
    
    plt.xlabel('$h_\mathrm{T}$ cut [GeV]', fontsize=14)
    plt.ylabel('$Z_N$', fontsize=14)
    
    ax.legend(loc='lower left', fancybox=True, framealpha=0.8)
 
    props = dict(boxstyle='round', facecolor='grey', alpha=0.5)
    ax.text(0.05, 0.90, ' '.join(['Stop mass =', str(mass), 'GeV']),
            bbox=props, transform=ax.transAxes, fontsize=12)

# draw the Zn for each mass
for mass in xrange(500, 1100, 100):
    plotZn(result_df, mass = mass)
    plt.savefig(''.join(['zn_vs_ht_cut.', str(mass), '.pdf']),
                bbox_inches='tight', format='pdf')
```

From these plots, it appears that the $m_{b\ell} > 200$ GeV region can be dropped. For the masses included in this study, this region never does better than the $m_{b\ell} > 400$ GeV region! Regarding the $m_{b\ell} > 800$ GeV region, it is tempting to say that performs significantly better than the $m_{b\ell} > 600$ GeV region, for the very high mass stops (900 and 1000 GeV). This is a little deceving though as the background estimate becomes a lot harder there.

### $Z_{N}$ vs $h_T$ cut value

```python
def plotBkgEst(df):
    """
    Function to take a data frame (df), and draw a plot of
    expected background events vs the hT cut value for
    different choices of mbl cut value. Each mbl cut value
    is a line on the plot
    """
    # the background information is repeated in each row, so
    # only grab the first sample
    first_sample = df.sample[0]
    
    # Create a copy of the data frame, with only the first sample
    # we can drop several columns while we are at it
    bkg = df[df.sample == first_sample][['ht cut',
                                         'mbl cut',
                                         'bkg total',
                                         'bkg uncertainty']]
    
    # get a list of the mbl cut values included in the data frame
    mbl_cut_vals = bkg['mbl cut'].unique()
    
    # list of new column names to add to the data frame used
    # for plotting
    new_cols = ['ht cut']
    new_cols.extend([' '.join(['$m_{bl}$ >',
                               str(mbl),
                               'GeV']) for mbl in mbl_cut_vals])
    estimate_to_print = pandas.DataFrame(columns=new_cols)
    uncertainty_to_print = pandas.DataFrame(columns=new_cols)
    rel_uncertainty_to_print = pandas.DataFrame(columns=new_cols)
    
    # step through the ht cut values, and add a new row
    # for each cut value
    for ht in bkg['ht cut'].unique():
        to_add_est = [ht]
        est_values = [bkg[(bkg['ht cut'] == ht) &
                               (bkg['mbl cut'] == mbl)].iloc[0]['bkg total']
                      for mbl in mbl_cut_vals]
        to_add_est.extend(est_values)

        to_add_uncert = [ht]
        uncert_values = [bkg[(bkg['ht cut'] == ht) &
                                  (bkg['mbl cut'] == mbl)].iloc[0]['bkg uncertainty']
                         for mbl in mbl_cut_vals]
        to_add_uncert.extend(uncert_values)
        
        to_add_rel_uncert = [ht]
        rel_uncert_values = [uncert/est for uncert, est in zip(uncert_values, est_values)]
        to_add_rel_uncert.extend(rel_uncert_values)

        estimate_to_print.loc[estimate_to_print.shape[0]] = to_add_est
        uncertainty_to_print.loc[uncertainty_to_print.shape[0]] = to_add_uncert
        rel_uncertainty_to_print.loc[rel_uncertainty_to_print.shape[0]] = (to_add_rel_uncert)
    
    # set the ht cut column to be the index of the data
    # frame
    estimate_to_print = estimate_to_print.set_index('ht cut').sort()
    uncertainty_to_print = uncertainty_to_print.set_index('ht cut').sort()
    rel_uncertainty_to_print = rel_uncertainty_to_print.set_index('ht cut').sort()

    # draw plot of Zn vs cut value
    fig, axes = plt.subplots(nrows=1, ncols=3)
    fig.set_figwidth(20)
    fig.set_figheight(5)
    estimate_to_print.plot(ax=axes[0])
    uncertainty_to_print.plot(ax=axes[1])
    rel_uncertainty_to_print.plot(ax=axes[2])

    
    [a.set_xlabel('$h_\mathrm{T}$ cut [GeV]', fontsize=14) for a in axes]
    axes[0].set_ylabel('Number events', fontsize=14)
    axes[1].set_ylabel('Uncertainty', fontsize=14)
    axes[2].set_ylabel('Uncertainty/Number events', fontsize=14)
    
    [a.set_yscale('log') for a in axes]
    axes[2].set_yscale('linear') # set this axis to linear by hand
    [a.legend(loc='best', fancybox=True, framealpha=0.8)]
     
    props = dict(boxstyle='round', facecolor='grey', alpha=0.5)
    axes[0].text(0.05, 0.90, 'Background prediction',
                 bbox=props, transform=axes[0].transAxes,
                 fontsize=12)
    axes[1].text(0.05, 0.90, 'Background uncertainty',
                 bbox=props, transform=axes[1].transAxes,
                 fontsize=12)
    axes[2].text(0.55, 0.85, 'Relative background\n uncertainty',
                 bbox=props, transform=axes[2].transAxes,
                 fontsize=12)

# Draw the background estimate!
plotBkgEst(result_df)
plt.savefig('bkg_est.pdf', bbox_inches='tight', format='pdf')
```

This suggests that the background estimate in the $m_{b\ell} > 800$ GeV region is very low with a large uncertainty (mostly driven by statistics). By moving to the $m_{b\ell} > 600$ GeV region we don't lose much expected sensitivity, but the background estimate should be more reliable and well understood.
